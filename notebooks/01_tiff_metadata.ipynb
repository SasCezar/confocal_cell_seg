```python
import tifffile

# Open TIFF file
with tifffile.TiffFile("/home/sasce/PycharmProjects/confocal_cell_seg/data/Project.lif - Series002.tif") as tif:
    # Print general info
    print("Number of pages:", len(tif.pages))

    # Access the first page
    page = tif.pages[0]

    # Standard TIFF tags
    print("Image shape:", page.shape)
    print("Data type:", page.dtype)
    print("Resolution:", page.tags.get("XResolution"))
    print("Software:", page.tags.get("Software"))

    # Iterate over all tags
    for tag in page.tags.values():
       name, value = tag.name, tag.value
       print(f"{name}: {value}")

    # # OME-XML or ImageJ metadata (if present)
    #if tif.ome_metadata:
    #     print("OME metadata:", tif.ome_metadata[:500])  # preview
    #if tif.imagej_metadata:
    #     print("ImageJ metadata:", tif.imagej_metadata)
```
